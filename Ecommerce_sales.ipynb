```python
df['Order Date']=pd.to_datetime(df['Order Date'])
print(df['Order Date'].dtype)
```

```python
# تجميع البيانات حسب اسم العميل وجمع مبيعات كل واحد، وبعدين ترتيبهم من الأكبر للأصغر
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

print("--- أفضل 10 عملاء من حيث إجمالي المبيعات ---")
print(top_customers)
```

```python
# تجميع أفضل 10 عملاء
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

# إعداد شكل الرسمة
plt.figure(figsize=(12, 6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='Blues_r')

# إضافة العناوين
plt.title('أعلى 10 عملاء من حيث إجمالي المبيعات')
plt.xlabel('إجمالي المبيعات')
plt.ylabel('اسم العميل')

plt.show()
```

```python
# 1. المبيعات حسب المنطقة (Region)
region_analysis = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

# 2. أفضل 5 مدن من حيث الأرباح (Profit)
top_cities_profit = df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(5)

print("--- إجمالي المبيعات لكل منطقة ---")
print(region_analysis)
print("\n--- أفضل 5 مدن تحقيقاً للأرباح ---")
print(top_cities_profit)
```

```python
# تجميع المبيعات حسب المنطقة
region_sales = df.groupby('Region')['Sales'].sum()

# إعداد الرسمة الدائرية
plt.figure(figsize=(8, 8))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))

plt.title('توزيع المبيعات حسب المنطقة')
plt.show()
```

```python
# تجميع أفضل 5 مدن في الأرباح
top_cities_profit = df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(5)

# رسم بياني بالأعمدة
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities_profit.index, y=top_cities_profit.values, palette='magma')

plt.title('أعلى 5 مدن تحقيقاً للأرباح')
plt.xlabel('المدينة')
plt.ylabel('إجمالي الأرباح')

plt.show()
```

```python
# تجميع الأرباح حسب فئة المنتج
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

print("--- الأرباح حسب الفئة ---")
print(category_profit)

```

```python
# 1. تجميع الأرباح حسب الفئة
cat_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

# 2. إعداد الرسم البياني
plt.figure(figsize=(10, 6))
# استخدام لون 'viridis' بيدي تدرج احترافي
sns.barplot(x=cat_profit.index, y=cat_profit.values, palette='viridis')

# 3. إضافة اللمسات النهائية
plt.title('إجمالي الأرباح لكل فئة من المنتجات')
plt.xlabel('الفئة')
plt.ylabel('إجمالي الأرباح')
plt.grid(axis='y', linestyle='--', alpha=0.7) # إضافة خطوط خفيفة عشان تقرأ الأرقام أسهل

plt.show()
```

```python
# 1. تجميع الأرباح حسب الفئة الفرعية
sub_cat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)

# 2. إعداد الرسم البياني
plt.figure(figsize=(12, 6))
sns.barplot(x=sub_cat_profit.values, y=sub_cat_profit.index, palette='rocket')

# 3. العناوين
plt.title('أعلى 10 فئات فرعية من حيث الأرباح')
plt.xlabel('إجمالي الأرباح')
plt.ylabel('الفئة الفرعية')

plt.show()
```

```python
# 1. حساب عدد مرات استخدام كل طريقة دفع
payment_counts = df['Payment Mode'].value_counts()

# 2. رسم دائرية (Pie Chart)
plt.figure(figsize=(8, 8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'))

plt.title('توزيع طرق الدفع المستخدمة')
plt.show()
```

```python
# رسم العلاقة بين الخصم والربح
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=df, hue='Category', alpha=0.7)

plt.title('العلاقة بين نسبة الخصم وإجمالي الأرباح')
plt.xlabel('الخصم (Discount)')
plt.ylabel('الأرباح (Profit)')
plt.axhline(0, color='red', linestyle='--') # خط يوضح نقطة الصفر (تحته خسران)

plt.show()
```

```python
# اختيار الأعمدة الرقمية فقط
numeric_cols = df[['Sales', 'Quantity', 'Discount', 'Profit']]

# حساب مصفوفة الارتباط
corr_matrix = numeric_cols.corr()

# رسم الـ Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('مصفوفة الارتباط بين المتغيرات الرقمية')
plt.show()
```

```python
# 1. استخراج الشهر من التاريخ (بايثون هيطلع أرقام من 1 لـ 12)
df['Month'] = df['Order Date'].dt.month

# 2. تجميع المبيعات حسب الشهر
monthly_sales = df.groupby('Month')['Sales'].sum()

# 3. رسم خط بياني (Line Chart) عشان نشوف التطور الزمني
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')

# 4. تظبيط الرسمة
plt.title('إجمالي المبيعات الشهرية خلال السنة')
plt.xlabel('الشهر')
plt.ylabel('إجمالي المبيعات')
plt.xticks(range(1, 13)) # التأكد إن كل الشهور ظاهرة على المحور
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
